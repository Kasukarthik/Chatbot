```python
import speech_recognition as sr 
import pyttsx3
import random
import pywhatkit
import datetime
import wikipedia
```

```python
listener=sr.Recognizer()
engine=pyttsx3.init()
voices=engine.getProperty('voices')
engine.setProperty('voice',voices[1].id)
def speaker(text):
    engine.say(text)
    engine.runAndWait()
    

def speak():
    try:
        with sr.Microphone() as source :
            print('listening...')
            voice=listener.listen(source,phrase_time_limit=6)
            command=listener.recognize_google(voice)
            command=command.lower()
            if ' siri ' in command:
                command=command.replace("siri"," ")
                #print(command)
                #speaker(command)    
    except:
        reply=['say again','come again','cant get you','please repeat','cant find you']
        speaker(random.choice(reply))
    return command
    
```
